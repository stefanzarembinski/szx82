```python
torch.autograd.set_detect_anomaly(False) 

def trainer():
    training_shell = TRAIN(
        file_name='bert',
        data_store=DATA_STORE,
        data=data(
            path.join(
            DATA_STORE, 
            'dev_tokenizer_piecewise_short;mean_len-15;seg_size-10;idx_step-1;level-4;_pre.pkl'), 
            DEVICE),
        model=PRE.MODEL,
    )

    # # ##### PROJECT
    training_shell.batch_size = 64
    training_shell.stop_thd=0.2
    training_shell.init(DEVICE, name_prep='dev', force=True)
    training_shell.train()

    '''DEFAULTS:
    '''
trainer()
```
